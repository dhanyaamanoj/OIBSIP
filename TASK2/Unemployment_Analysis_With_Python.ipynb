## Visualizing The Dataset

```python
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
plt.show
```

```python
plt.figure(figsize=(10,10))
plt.title("Unemployment in india")
sns.histplot(x=' Estimated Unemployment Rate (%)',hue= "Region", data=df,kde=False)
plt.show()
```

```python
#Barplot -Estimated Employed V/S Region

plt.figure(figsize=(10,10))
sns.barplot(x=' Estimated Employed',y= "Region", data=df)
plt.show()
```

The highest number of people employed in UP and the lowest is Sikkim

```python
plt.figure(figsize=(10,10))
sns.barplot(x=' Estimated Unemployment Rate (%)',y= "Region", data=df)
plt.show()
```

The highest unemployment is in Haryana,while the lowest is in Meghalaya

```python
#Barplot-Estimated Labour Participation Rate V/s Region
plt.figure(figsize=(10,10))

sns.barplot(x=' Estimated Labour Participation Rate (%)',y= "Region", data=df)
plt.show()
```

Estimate labour participation is high in Meghalaya whereas its comparatively low in Kerala.This might be the reason why Meghalaya has lowe unemployment rate as compared to other states.
